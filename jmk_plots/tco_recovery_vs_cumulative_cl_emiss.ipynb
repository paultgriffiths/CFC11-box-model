```python
import pylab as plt
import numpy as np
import pandas as pd
import matplotlib as mpl
from scipy import stats
import numpy as np
```

```python
mpl.rc('figure', figsize=(15,15),facecolor='white', dpi=300)
mpl.rc('figure.subplot', wspace= 0.4, hspace=0.2)
mpl.rc('figure.subplot', hspace= 0.2)
mpl.rc('axes', grid=1, titlesize=14)
mpl.rc('axes.formatter',limits=(-2,5))
mpl.rc('lines', linewidth=2, color='r')
mpl.rc('font', size=16)
```

```python
data = pd.read_csv("tco_recovery_vs_cl_emiss.csv")#, header=0, sep=',')
slope, intercept, r_value, p_value, std_err = stats.linregress(data['Emiss'],data['Year'])
```

```python
colors4plot = ['r', 'b', 'g', 'cyan', 'orange', 'yellow', 'g', 'cyan', 'orange', 'yellow' ]
```

```python
plt.figure(figsize=(15,15),dpi=200)
plt.clf()

for ivar in range(0,10):
    plt.plot(data['Emiss'][ivar], data['Year'][ivar], 'o', 
             color=colors4plot[ivar], label=data['Experiment'][ivar])
plt.legend(ncol=2, fontsize=8)

plt.plot([data['Emiss'][0], data['Emiss'][9]],[intercept+slope*data['Emiss'][0],intercept+slope*data['Emiss'][9]])

plt.figtext(x=0.2,y=0.2,s="phallus dei", fontsize=6)
```
